```python
import numpy as np
import matplotlib.pyplot as plt


data = np.fromfile("drive/MyDrive/CUDA/data.dat", dtype=np.float32)
tDim = 20000
xDim = 100
yDim = 500
data = np.reshape(data,(tDim,yDim,xDim))

activity = np.fromfile("drive/MyDrive/CUDA/activity.dat", dtype=np.float32)
tDim = 20000
xDim = 100
yDim = 500
activity = np.reshape(activity,(tDim,yDim,xDim))
```

```python
#activity = activity/np.amax(activity)
print(activity.shape)
oxideConcentration = np.sum(activity,axis=0)
print(oxideConcentration.shape)
plt.figure()
plt.imshow(oxideConcentration)
plt.colorbar()
plt.title("Oxide concentration")
plt.show()

plt.figure()
plt.plot(oxideConcentration[50,:])
plt.title("Oxide concentration Slice")
plt.show()

for x in range(50):
  plt.subplot(1, 2, 1)
  plt.imshow(data[x*tDim//50,:,:], vmin = 0, vmax=1)
  plt.colorbar()
  plt.title("Oxygen Concentration")

  plt.subplot(1, 2, 2)
  plt.imshow(activity[x*tDim//50,:,:], vmin = 0)
  plt.colorbar()
  plt.title("Activity")
  plt.show()

plt.figure()

plt.plot(data[:,10,10])
plt.plot(data[:,20,20])
plt.plot(data[:,30,30])
plt.plot(data[:,40,40])
plt.plot(data[:,50,50])
plt.show()

plt.figure()

plt.plot(data[-1,10,:])
plt.plot(data[-1,20,:])
plt.plot(data[-1,30,:])
plt.plot(data[-1,40,:])
plt.plot(data[-1,50,:])
plt.show()
```
